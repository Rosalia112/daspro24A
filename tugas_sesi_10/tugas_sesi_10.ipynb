```python
import numpy as np
import pandas as pd
from datetime import datetime

n_data = 1000

kota = [f"Kota{str(i).zfill(4)}" for i in range(1, n_data + 1)]
tgl = [datetime.now().strftime("%Y-%m-%d %H:%M")for _ in range(n_data)]
musim = np.random.choice(['panas','hujan'], size=n_data)
kc = np.random.choice(['cerah','hujan','hujan lebat','berawan','hujan disertai petir','berangin','mendung'] , size=n_data)
libur = np.random.choice(['Ya' , 'Tidak'] , size=n_data)
bhy_cuaca = np.random.choice(['Berbahaya' , 'Aman'] , size=n_data)
suhu_c = np.random.randint(0,45, size=n_data)
suhu_f = np.random.randint(50,113, size=n_data)
uv_i = np.random.randint(0,11, size=n_data)
klmb = np.random.randint(20,90, size=n_data)
c_hujan = np.random.randint(0,120, size=n_data)
k_angin = np.random.randint(5,30, size=n_data)



df = pd.DataFrame({
    'Kota :' : kota ,
    'Tanggal dan waktu :' : tgl ,
    'Musim :' : musim ,
    'Kondisi Cuaca :' : kc ,
    'Apakah kota sedang libur? :' : libur ,
    'Status Cuaca :' : bhy_cuaca ,
    'Suhu kota (Celcius) :' : suhu_c ,
    'Suhu kota (Fahrenheit) :' : suhu_f ,
    'UV indeks :' : uv_i ,
    'Kelembaban(%) :' : klmb ,
    'Curah hujan(mm/hari) :' : c_hujan ,
    'Kecepatan angin(km/jam) :' : k_angin

})

print(df)
print(df.info)
print(df.head())
print(f"Kondisi cuaca yang paling sering muncul adalah : {df['Kondisi Cuaca :'].value_counts().idxmax()}")
print("====== Informasi Detail Suhu Kota")
print(f"Suhu terkecil : {df['Suhu kota (Celcius) :'].min()} Celcius")
print(f"Suhu terkecil berada di kota : {df.loc[df['Suhu kota (Celcius) :'].min() , 'Kota :']} ")
print(f"Suhu terbesar : {df['Suhu kota (Celcius) :'].max()} Celcius")
print(f"Suhu terbesar berada di kota : {df.loc[df['Suhu kota (Celcius) :'].max() , 'Kota :']} ")
print("====== Informasi detailUV Indeks ======")
print(f"UV indeks terkecil : {df['UV indeks :'].min()}")
print(f"UV indeks terkecil berada di kota : {df.loc[df['UV indeks :'].min() , 'Kota :']} ")
print(f"UV indeks terbesar : {df['UV indeks :'].max()}")
print(f"UV indeks terbesar berada di kota : {df.loc[df['UV indeks :'].max() , 'Kota :']} ")
print("===========================================================")
jml_libur = df[df['Apakah kota sedang libur? :']=='Ya']
print(f"Jumlah Kota yg sedang libur : {jml_libur.size/12}")
print("===========================================================")
print(f"Curah Hujan terkecil : {df['UV indeks :'].min()}")
print(f"Curah Hujan terkecil berada di kota : {df.loc[df['Curah hujan(mm/hari) :'].min() , 'Kota :']} ")
print(f"Curah Hujan terbesar : {df['UV indeks :'].max()}")
print(f"Curah hujan terbesar berada di kota : {df.loc[df['Curah hujan(mm/hari) :'].max() , 'Kota :']} ")
print("===========================================================")
jml_hujan_lebat = df[df['Kondisi Cuaca :']=='hujan lebat']
print(f"Jumlah Kota yang sedang Hujan Lebat : {jml_hujan_lebat.size/12}")
print("===========================================================")
jml_hujan_petir = df[df['Kondisi Cuaca :']=='hujan disertai petir']
print(f"Jumlah Kota yang sedang Hujan disertai petir : {jml_hujan_petir.size/12}")

```
